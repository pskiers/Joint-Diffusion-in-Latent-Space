```python
import torch

# Load the checkpoint
checkpoint_path = '/home/jk/Joint-Diffusion-in-Latent-Space/models_imagenet/imagenet_diff.cpkt'
checkpoint = torch.load(checkpoint_path)

# Function to rename the keys
def rename_keys(state_dict):
    new_state_dict = {}
    for key in state_dict.keys():
        if "model_ema" in key:
            new_key = key.replace('diffusion_', 'modeldiffusion_')
            new_state_dict[new_key] = state_dict[key]
        else:
            new_state_dict[key] = state_dict[key]
    return new_state_dict


def squeeze_keys(state_dict):
    new_state_dict={}
    for key in state_dict.keys():
        #model_ema.modeldiffusion_modelinput_blocks11proj_outweight
        if "model_ema" in key and "block" in key:
            print(key)
            new_state_dict[key] = state_dict[key].squeeze(-1)
        else:
            new_state_dict[key] = state_dict[key]
    return new_state_dict


checkpoint['state_dict'] = rename_keys(checkpoint['state_dict'])
#checkpoint['state_dict'] = squeeze_keys(checkpoint['state_dict'])

# Save the updated checkpoint
new_checkpoint_path = '/home/jk/Joint-Diffusion-in-Latent-Space/models_imagenet/imagenet_diff_renamed.cpkt'
torch.save(checkpoint, new_checkpoint_path)

print(f"Checkpoint saved to {new_checkpoint_path}")
```

```python
import torch

# Load the checkpoints
checkpoint_path = '/home/jk/Joint-Diffusion-in-Latent-Space/models_imagenet/imagenet_diff_renamed.cpkt'
ref_checkpoint_path = '/home/jk/Joint-Diffusion-in-Latent-Space/models_imagenet/ref_weights.ckpt'

checkpoint = torch.load(checkpoint_path)
ref_checkpoint = torch.load(ref_checkpoint_path)

# Reshape the weights
for key in checkpoint['state_dict']:
    if key in ref_checkpoint['state_dict']:
        print(key)
        # Get the shapes of the weights
        checkpoint_shape = checkpoint['state_dict'][key].shape
        ref_checkpoint_shape = ref_checkpoint['state_dict'][key].shape

        if checkpoint_shape != ref_checkpoint_shape:
            print(f'Reshaping {key} from {checkpoint_shape} to {ref_checkpoint_shape}')
            checkpoint['state_dict'][key] = checkpoint['state_dict'][key].reshape(ref_checkpoint_shape)

# Optionally save the modified checkpoint
modified_checkpoint_path = '/home/jk/Joint-Diffusion-in-Latent-Space/models_imagenet/imagenet_diff_renamed_modified.cpkt'
torch.save(checkpoint, modified_checkpoint_path)
```
